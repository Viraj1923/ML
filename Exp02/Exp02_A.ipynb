```python
 import pandas as pd
 import numpy as np
```

```python
df=pd.read_csv('sales_data_3.csv')
df
```

```python
# What is the average age of the customers?
```

```python
avg_age=df['Customer_Age'].mean()
avg_age
```

```python
# How do different age groups behave in terms of purchases?
```

```python
behave=df.groupby('Age_Group')['Revenue'].sum()
behave
```

```python
 # What are the purchasing patterns across different genders
```

```python
pur_pattern=df.groupby('Customer_Gender')['Revenue'].sum()
pur_pattern
```

```python
# Which countries/states have the highest/lowest sales revenue
```

```python
sale_revenue_c=df.groupby('Country')['Revenue'].sum()
high=sale_revenue_c.idxmax()
low=sale_revenue_c.idxmin()
print(high)
print(low)
```

```python
sale_revenue_s=df.groupby('State')['Revenue'].sum()
high_s=sale_revenue_s.idxmax()
low_s=sale_revenue_s.idxmin()
print(high_s)
print(low_s)
```

```python
# Are there geographical trends in product preference?
```

```python
# Do customers from certain countries/states tend to buy certain        products?
```

```python
coun_tend=df.groupby(['Country','Product'])['Revenue'].sum()
coun_tend
```

```python
# Which products/sub-categories have the highest sales/revenue?
```

```python
p_max=df.groupby('Product')['Revenue'].sum().idxmax()
sp=df.groupby('Sub_Category')['Revenue'].sum().idxmax()
print(p_max)
print(sp)
```

```python
# What is the profitability of different products/categories?
```

```python
profitability=df.groupby(['Product'])['Profit'].sum()
profitability
```

```python
# Which products have the highest order quantities?
```

```python
high_order=df.groupby('Product')['Order_Quantity'].sum().idxmax()
high_order
```

```python
# Which products generate the highest profit margins?
```

```python
df['Profit_margin'] = df['Profit'] / df['Revenue']
df
```

```python
high_profit_margins=df.groupby('Product')['Profit_margin'].sum().idxmax()
high_profit_margins
```

```python
# Are the seasonal trends in product purchases based on the date/month/year?
```

```python
df.info()
```

```python
df['Date']=pd.to_datetime(df['Date'])
df.info()
```

```python
df['Month']=df['Date'].dt.month
df['Year']=df['Date'].dt.year
df
```

```python
# What is the distribution of order quantities across different products?
```

```python
order_quantity_distribution = df.groupby('Product')['Order_Quantity'].sum().describe()
print(order_quantity_distribution)
```

```python
# Which products are often purchased in bulk?
```

```python
bulk_purchases = df[df['Order_Quantity'] > 10].groupby('Product')['Order_Quantity'].sum().idxmax()
print(bulk_purchases)
```

```python
# What is the total revenue and profit generated by each product/category?
```

```python
revenue_profit_by_category = df.groupby(['Product_Category', 'Product'])[['Revenue', 'Profit']].sum()
print(revenue_profit_by_category)
```
