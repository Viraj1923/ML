```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df=pd.read_csv("advertising.csv")
df.head()
```

```python
#independent
x=df.drop(columns=["Sales"])
print(x)
```

```python
y=df["Sales"]
print(y)
```

```python
df.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
reg_model=LinearRegression()
```

```python
reg_model.fit(x_train,y_train)
```

```python
predictions=reg_model.predict(x_test)
print(predictions)
```

```python
#r2

score=r2_score(y_test,predictions)
print(score)
```

```python
#mean_square
sc=mean_squared_error(y_test,predictions)
print(sc)
```

```python
#coefficeint
coe=reg_model.coef_
print(coe)
```

```python
#intercept
inter=reg_model.intercept_
print(inter)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
df = pd.read_excel("Concrete_Data.xls")
df
```

```python
x = df.drop(columns = ["Concrete compressive strength(MPa, megapascals) "])
print(x)
```

```python
y = df["Concrete compressive strength(MPa, megapascals) "]
y
```

```python
#train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
#train the model
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
predications = model.predict(X_test)
print(predications)
```

```python
#Mean Squared Error(MSE)
print("mean_squared_error : ",mean_squared_error(Y_test,predications))
```

```python
#R2(Variance)
accuracy = r2_score(Y_test,predications)
print("Accuracy is : ",accuracy)
```
